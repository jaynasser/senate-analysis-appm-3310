```python
import os

from dotenv import load_dotenv
import pandas as pd
```

```python
load_dotenv()

SENATORS_CSV_FP = os.environ['SENATORS_CSV_FP']
VOTES_CSV_FP = os.environ['VOTES_CSV_FP']
```

```python
df_senators = pd.read_csv(SENATORS_CSV_FP)
df_votes = pd.read_csv(VOTES_CSV_FP)
```

```python
df_senators.party_code.replace(328, 100, inplace=True) # classify independents as democrats for simplicity (Sanders and King)

df_senators.drop(df_senators[df_senators.icpsr == 99912].index, inplace=True) # Trump

df_senators = df_senators[['icpsr', 'state_abbrev', 'party_code', 'bioname', 'nominate_number_of_votes']]
```

```python
df_votes.drop(['congress', 'chamber', 'prob'], axis=1, inplace=True)

df_votes.drop(df_votes[df_votes.icpsr == 99912].index, inplace=True) # Trump
```

```python
df_all = df_votes.merge(df_senators, how='left', on='icpsr')

df_all.set_index(['rollnumber', 'icpsr'], inplace=True)
```
